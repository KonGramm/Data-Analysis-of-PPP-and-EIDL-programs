Task 1: Load into an indexed Pandas data frame the variables: Loan Amount, Borrower’s State, Date Approved and NAICS Code from all available PPP files (keep only two-digit NAICS)

```python
# map variable types to later be correctly loaded
dtype_mapping = {'DateApproved'         : 'str', 
                 'BorrowerState'        : 'category',
                 'CurrentApprovalAmount': 'float32', 
                 'NAICSCode'            : 'str'
                }
```

```python
# select appropriate columns for analysis
selected_columns = ['DateApproved', 'BorrowerState', 'CurrentApprovalAmount', 'NAICSCode']
```

```python
# Define the folder path where my PPP data are

folder_path = 'C:\\Users\\kosti\\OneDrive\\Desktop\\Data Engineering\\Homework2\\CovidRecovery\\RawData'
output_folder = 'C:\\Users\\kosti\\OneDrive\\Desktop\\Data Engineering\\Homework2\\CovidRecovery\\RawData'


def process_file(file_path, output_path):
    
    # Load the data with specified data types and only selected columns
    df = pd.read_csv(file_path, usecols=selected_columns, dtype=dtype_mapping)

    df['NAICSCode'] = df['NAICSCode'].str[:2].astype('category')

    # Save the sampled data of each file to a new CSV file
    df.to_csv(output_path, index=False)
    
    return df
```

```python
# create an empty list to fill with each file
results = []

# list of files to work with
csv_files = [f"public_up_to_150k_{i}_230930.csv" for i in range(1, 13)]
csv_files = [os.path.join(folder_path, f) for f in csv_files]
```

```python
# process each file
for file_path in csv_files:
    output_path = os.path.join(output_folder, f'sampled_{os.path.basename(file_path)}')
    df = process_file(file_path, output_path)
    results.append(df)

# Concatenate all DataFrames in the results list into a single DataFrame
combined_df = pd.concat(results)
```

```python
import geopandas as gpd
## Make the map plot

# Path to the shapefiles
shapefile_path = 'C:\\Users\\kosti\\OneDrive\\Desktop\\Data Engineering\\Homework2\\CovidRecovery\\RawData\\cb_2020_us_state_20m.shp'

# Load the US states shapefile
gdf = gpd.read_file(shapefile_path)

# Rename 'STUSPS' to 'State'
gdf = gdf.rename(columns={'STUSPS': 'STATE'})

print("Columns in gdf:", gdf.columns)
print("Columns in merged_data:", merged_data.columns)


# Merge the geospatial data with the normalized loan data
# 'STATE' is the common column for merging
merged_gdf = gdf.set_index('STATE').join(merged_data.set_index('State'))


print("Columns in merged gdf:", merged_gdf.columns)

```

```python
# Only plot for 'NormalizedLoanNumber'
column_to_plot = 'NormalizedLoanNumber'

# Create heat map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
contiguous_usa = merged_gdf[~merged_gdf.index.isin(['AK', 'HI', 'PR'])]  # Exclude Alaska, Hawaii, and Puerto Rico
contiguous_usa.plot(column=column_to_plot, cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True,vmin=0, vmax=1)
ax.set_title(f'Heat Map of {column_to_plot.replace("_", " ")}')
ax.axis('off')

# Add labels to the states
for idx, row in contiguous_usa.iterrows():
    ax.annotate(text=idx, xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                horizontalalignment='center', fontsize=8, color='white', weight='bold')

plt.show()
```
